```python
def initDF(path):
    df = pd.read_csv('/kaggle/input/'+path)
    return df
```

```python
df_LA = initDF("crimes-in-los-angeles-2020-2023/Crime_Data_from_2020_to_Present.csv")
df_LA.head()
```

```python
df_Phil = initDF("philadelphiacrimedata/crime.csv")
df_Phil.head()
```

```python
df_Bal = initDF("crime-in-baltimore/BPD_Part_1_Victim_Based_Crime_Data.csv")
df_Bal.head()
```

```python
df_Port = pd.read_csv('/kaggle/input/portland-crime/output.csv', delimiter='\t')
df_Port.head()
```

```python
def checkNull(df):
    return df.isnull().sum()
```

```python
dfs = ["df_LA", "df_Phil", "df_Bal", "df_Port"]
```

```python
df_LA = df_LA[["date_occurred", "crime_description", "location", "latitude", "longitude"]]
df_LA.head()
```

```python
checkNull(df_LA)
```

```python
df_LA[['date','time']] = df_LA.date_occurred.str.split(expand=True)
df_LA.head()
```

```python
df_LA['city'] = 'Los Angeles'
df_LA.head()
```

```python
df_LA = df_LA[["date", "time", "crime_description", "location", "latitude", "longitude", "city"]]
df_LA.head()
```

```python
df_Bal = df_Bal[["CrimeDate", "CrimeTime", "Description", "Location", "Latitude", "Longitude"]]
df_Bal["City"] = "Baltimore"
df_Bal.head()
```

```python
df_Bal = df_Bal.dropna()
```

```python
checkNull(df_Bal)
```

```python
checkNull(df_Phil)
```

```python
df_Phil = df_Phil[["Dispatch_Date", "Dispatch_Time", "Text_General_Code", "Location_Block", "Lat", "Lon"]]
df_Phil["City"] = "Philadephia"
df_Phil.head()
```

```python
df_Phil = df_Phil.dropna()
checkNull(df_Phil)
```

```python
checkNull(df_Port)
```

```python
df_Port = df_Port[["OccurDate", "OccurTime", "Address", "OffenseType", "OpenDataLat", "OpenDataLon"]]
df_Port["City"] = "Portland"
df_Port
```

```python
df_Port = df_Port.dropna()
df_Port
```

```python
checkNull(df_Port)
```

```python
df_LA.rename(columns={"date": "Date", "time": "Time", "crime_description": "Description", 
                      "location": "Location", "latitude": "Latitude", "longitude": "Longitude", 
                      "city": "City"}, inplace=True)
df_LA.head()
```

```python
df_Bal.rename(columns={"CrimeDate": "Date", "CrimeTime": "Time"}, inplace=True)
df_Bal.head()
```

```python
df_Phil.rename(columns={"Dispatch_Date": "Date", "Dispatch_Time": "Time", "Text_General_Code": "Description", 
                        "Location_Block": "Location", "Lat": "Latitude", "Lon": "Longitude"}, inplace=True)
df_Phil.head()
```

```python
df_Port.rename(columns={"OccurDate": "Date", "OccurTime": "Time", "Address": "Location", "OffenseType": "Description", 
                        "OpenDataLat": "Latitude", "OpenDataLon": "Longitude"}, inplace=True)
df_Port = df_Port[["Date", "Time", "Description", "Location", "Latitude", "Longitude", "City"]]
df_Port.head()
```

```python
df_Port.tail()
```

```python
# for tm in df_Port["Time"]:
#     df_Port["Time"] = str(tm)[:-2] + ":" + str(tm)[-2:]
    

def chgTime(tm):
    return str(tm)[:-2] + ":" + str(tm)[-2:]

df_Port["Time"] = df_Port["Time"].apply(chgTime)
df_Port
```

```python
df_Bal[['Month', 'Day', 'Year']] = df_Bal.Date.str.split('/', expand=True)
df_Bal.head()
```

```python
df_Bal = df_Bal.drop("Date", axis=1)
df_Bal.head()
```

```python
df_Bal["Date"] = df_Bal["Year"] + "-" + df_Bal["Month"] + "-" + df_Bal["Day"]
df_Bal.head()
```

```python
df_Bal = df_Bal[["Date", "Time", "Description", "Location", "Latitude", "Longitude", "City"]]
df_Bal.head()
```
